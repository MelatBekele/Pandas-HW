## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# creating bins 
bins_age = [0,9,14,19,24,29,34,39,60]
grouping= ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#pd.cut()
#purchase_data= pd.cut()

Players = purchase_data[["SN", "Age"]]
Players_cout= Players['SN'].nunique()

Players["Age group"] = pd.cut(Players["Age"], bins_age, labels=grouping)

#Players.loc[row_indexer,col_indexer

age_grouping = Players.groupby("Age group")
age_grouping

#age_grouping_count = Players["SN"].nunique()
age_grouping_count = age_grouping["SN"].nunique()
age_grouping_count

Percentage_age= ((age_grouping_count/Players_cout)*100).round(2)
Percentage_age
```

```python
#dataframe

Demographics = pd.DataFrame({ "Total Count": age_grouping_count  , "Percentage of Players": Percentage_age})

Demographics["Percentage of Players"]= Demographics["Percentage of Players"].astype(float).map("${:}%".format)
Demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Purchase = purchase_data[["SN", "Age", "Purchase ID", "Price"]]
Purchase_cout= Purchase['Purchase ID'].nunique()


Purchase["Age group"] = pd.cut(Purchase["Age"], bins_age, labels=grouping)
Purchase_grouping = Purchase.groupby("Age group")
Purchase_grouping.head(10)

ava_pur = Purchase_grouping["Price"].mean()

Purchase_cout_age= Purchase_grouping["Purchase ID"].count()

pur_sum = Purchase_grouping["Price"].sum()
pur_sum 

ava_age_price = ava_pur_sum/ava_pur
ava_age_price 


```

```python
Purch_age = pd.DataFrame({"Purchase Count": Purchase_cout_age, "Average Purchase Price": ava_pur , "Total Purchase Value":pur_sum , "Avg PurchaseTotal per Person ":ava_age_price })
Purch_age

Purch_age ["Average Purchase Price"] = Purch_age["Average Purchase Price"].astype(float).map("${:,.2f}".format)
Purch_age ["Total Purchase Value"] = Purch_age["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Purch_age ["Avg PurchaseTotal per Person "] = Purch_age["Avg PurchaseTotal per Person "].astype(float).map("${:,.2f}".format)
Purch_age
```
